```python
import numpy as np
```

```python
def is_near_equal(A, B, tol=1e-5):
  """
  Checks if two matrices are element-wise near equal within a tolerance.

  Args:
      A: First matrix.
      B: Second matrix.
      tol: Tolerance level (default: 1e-5).

  Returns:
      True if matrices are near equal, False otherwise.
  """
  return np.abs(A - B).max() <= tol
```

```python
def isorthonormal(matrix):
  #check if the matrix is square
  rows, cols = matrix.shape
  if rows != cols:
    return False
  # Calculate eigenvalues and eigenvectors
  eigenvalues, eigenvectors = np.linalg.eig(matrix)

  # Check if eigenvectors are orthonormal
  eigenvector_dot_product = np.dot(eigenvectors.T, eigenvectors)
  print("Eigenvector Product Matrix: \n",eigenvector_dot_product)
  is_orthonormal = is_near_equal(np.eye(3),eigenvector_dot_product) # check the identity matrix(np.eye(3)) and dot product are same

  # Print the results
  print("\nEigenvalues:")
  print(eigenvalues)
  print("\nEigenvectors:")
  print(eigenvectors)
  print("\nAre eigenvectors orthonormal:", is_orthonormal)
```

```python
#define matrix
A = np.array([[6, -2, -1],
              [-2,  6, -1],
              [-1, -1,  5]])

isorthonormal(A)
```

***P, D, P and P^-1 Construction***

```python
def construct_p_d_pinv(matrix):
    # Calculate eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(matrix)

    # Construct D matrix (diagonal matrix of eigenvalues)
    D = np.diag(eigenvalues)

    # Calculate the inverse of P matrix
    P_inv = np.linalg.inv(eigenvectors)

    return eigenvectors, D, P_inv
```

```python
P, D, P1 = construct_p_d_pinv(A)
```

```python
print("\nP:\n",P)
print("\nD:\n",D)
print("\nP^-1:\n",P1)
```
