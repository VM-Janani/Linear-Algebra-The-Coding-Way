What does basis vectors mean ?

A basis is a set of vectors that are linearly independent and can be used to express any vector within its vector space as a linear combination of these basis vectors.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def plot_transformation(matrix,e1,e2,e1_transformed,e2_transformed,x,dx):

    # Plot original basis vectors
    plt.arrow(0, 0, e1[0], e1[1], color='b', width=0.05, head_width=0.2, head_length=0.2, length_includes_head=True, label='e1')
    plt.arrow(0, 0, e2[0], e2[1], color='r', width=0.05, head_width=0.2, head_length=0.2, length_includes_head=True, label='e2')

    # Plot transformed basis vectors
    plt.arrow(0, 0, e1_transformed[0], e1_transformed[1], color='b', linestyle='--', width=0.05, head_width=0.2, head_length=0.2, length_includes_head=True, label='e1 transformed')
    plt.arrow(0, 0, e2_transformed[0], e2_transformed[1], color='r', linestyle='--', width=0.05, head_width=0.2, head_length=0.2, length_includes_head=True, label='e2 transformed')
    plt.arrow(0, 0, x[0], x[1], color='g', width=0.05, head_width=0.2, head_length=0.2, length_includes_head=True, label='x')
    plt.arrow(0, 0, dx[0], dx[1], color='g', linestyle='--', width=0.05, head_width=0.2, head_length=0.2, length_includes_head=True, label='dx')



    # Set plot limits
    xlim_max = max(np.abs(e1_transformed[0]), np.abs(e2_transformed[0]))
    ylim_max = max(np.abs(e1_transformed[1]), np.abs(e2_transformed[1]))
    plt.xlim(-xlim_max, xlim_max)
    plt.ylim(-ylim_max, ylim_max)

    # Add labels and legend
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()

    # Show plot
    plt.title('Transformation of Basis Vectors')
    plt.show()
```

```python
# Define basis vectors
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Define transformation matrix tx (or dx)
D = np.array([[3, 0],
               [0, 6]])

# Transform basis vectors using tx
e1_transformed = np.dot(D, e1)
e2_transformed = np.dot(D, e2)

plot_transformation(D,e1,e2,e1_transformed,e2_transformed)
```

```python
# Define basis vectors
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Define transformation matrix tx (or dx)
D = np.array([[3, 0],
               [0, 6]])

x=[1,1]
dx=np.dot(D, x)
# Transform basis vectors using tx
e1_transformed = np.dot(D, e1)
e2_transformed = np.dot(D, e2)

plot_transformation(D,e1,e2,e1_transformed,e2_transformed,x,dx)
```
