**problem statement:**

In this problem we have an internet with 4 pages, namely page1, page2, page3, and page4, which refer to each other.Find the Rank of each pages in internet using linear algebra

**solution:**

Step 1:Construct a Directed Graph.

Step 2:Build a probabilistic matrix.

Step 3:formulate linear algebraic equation

Step 4:Solve the equation and get the rank as eigenvectors




**1.Construct a Directed Graph:**

**2.Build a probabilistic matrix:**

```python
import numpy as np
```

```python
A = np.array([[0, 0, 1, 1/2],
              [1/3, 0, 0, 0],
              [1/3, 1/2, 0, 1/2],
              [1/3, 1/2, 0, 0]])
```

 **3.Formulate linear algebraic equation**

 x1= x3+1/2 x4

 x2=1/3 x1

 x3=1/3 x1+1/2 x2+1/2 x4

 x4=1/3 x1+1/2 x2

x=[x1,x2,x3,x4]

x=Ax

solve the equtiona and find the rank:

Compute the eigenvalues and eigenvectors of the probability matrix .

Identify the eigenvector corresponding to the eigenvalue 1. This eigenvector represents the ranks of the pages.

Normalize the dominant eigenvector so that its elements sum to 1.




```python
# Calculate eigenvalues and eigenvectors of 'st_matrix_array'
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigen values are:")
print(eigenvalues)

print("\nEigen vectors:")
print(eigenvectors)
```

```python
# get the real part of eigenvalues
eigenvalue_real = np.real(eigenvalues)
print("\nClosest Eigen value considering only real part:")
print(eigenvalue_real)
```

this function used to find the eigenvector related with the real part of the eigen values

```python
def EigenVector_real(eigenvalue_real, eigenvectors):
    no= len(eigenvalue_real)

    # Loop through each eigenvalue
    for i in range(no):
        # Check if the eigenvalue is approximately equal to 1.0 after rounding
        if round(eigenvalue_real[i]) == 1.0:
            # Extract the corresponding eigenvector
            eigenvector_realpart = np.real(eigenvectors[:, i])
            print("Eigen vector corresponds to Eigen value 1:")
            print(eigenvector_realpart)
    # Scale the eigenvector by dividing each element by the sum
    sum_eigenvector = np.sum(eigenvector_realpart)
    scaled = eigenvector_realpart.real / sum_eigenvector
    print("the scaled value")
    print(scaled)
    #the sum of
    print("Sum of the scaled values",round(sum(scaled),6))


```

```python
EigenVector_real(eigenvalue_real, eigenvectors)
```
