```python
# Probability Matrix
P = [[0, 0, 1, 0.5],
     [1/3, 0, 0, 0],
     [1/3, 0.5, 0, 0.5],
     [1/3, 0.5, 0, 0]]
```

```python
def transition_probabilities(transition_matrix, initial_probabilities, iterations):
    """
    Simulates the transition probabilities of moving between pages over a certain number of iterations.

    Args:
    - transition_matrix (list of lists): The transition matrix representing the probabilities of moving between pages.
    - initial_probabilities (list): The initial probabilities of landing on each page.
    - iterations (int): The number of iterations to simulate.

    """
    # Initialize probabilities with initial probabilities
    probabilities = initial_probabilities.copy()

    # Iterate for the specified number of iterations
    for _ in range(iterations):
        # Initialize a new list to store updated probabilities
        updated_probabilities = []

        # Calculate new probabilities based on transition matrix
        for i in range(len(transition_matrix)):
            new_prob = 0
            for j in range(len(transition_matrix[i])):
                new_prob += transition_matrix[i][j] * probabilities[j]
            updated_probabilities.append(new_prob)

        # Update probabilities for the next iteration
        probabilities = updated_probabilities

    return probabilities
```

Let’s say you start on Page 1, and then depending on this matrix, you move to the next page. Repeat this process, say
a hundred times. How would the composition of vector x look ?

answer:

After 100 iterations, the composition of vector
x will represent the long-term probabilities of being on each page. These probabilities represent where you're likely to end up after moving through the pages for a hundred times, starting from Page 1.








```python
x1 = [1, 0, 0, 0] #probability of landing on other pages from page 1
n = 100
new_x1 = transition_probabilities(P, x1, n)

print("After", n, "iterations the probabilities of landing on each page are:\n", [round(prob, 4) for prob in new_x1])
```

If you start on Page 3 and repeat the process of moving through the pages according to the transition matrix for a hundred times, the composition of the vector
x would represent the long-term probabilities of being on each page.

This vector
x is called the steady-state probability vector or the stationary distribution of the Markov chain. It represents the long-term probabilities of being in each state (or on each page in this case) after the Markov chain has stabilized.
