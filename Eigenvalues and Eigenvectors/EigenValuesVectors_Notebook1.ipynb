1. Select the vectors and Perform the matrix transformation (by multiplying these vectors with the given matrix A).

2. Plot the line graph for the original vectors u and v, along with their transformed vectors Au and Av.

3. Analyze the graph: If the direction of a vector remains the same after transformation(it can be scaled), it is considered an eigenvector of the matrix. Conversely, if the direction changes, it is not an eigenvector.

Av=λv the vector can be scaled after matrix trasformation by a value of λ which an eigen value


```python
#import matplotlib
import matplotlib.pyplot as plt
```

we can use user defined function for built in numpy **dot** fuction for matrix multhiplication.


---



```python
#function to multiply the matrix and the vector
def mul_fun(a,b):
  if len(a[0])!=len(b):
    raise ValueError("no of column in matrix and vector lenght should be equal")
  result= []
  for i in range(len(a)):
        sum = 0
        for k in range(len(b)):
            sum += a[i][k] * b[k] #multiply the elements of the matrix and the vector
        result.append(sum)
  return result
```

```python
#function to plot the line graph
def plot_graph(A, u, v):
    #defining start and end points
    start1 = (0, 0)
    end1 = mul_fun(A, u)

    start2 = (0, 0)
    end2 = mul_fun(A, v)

    start3 = (0, 0)
    end3 = u

    start4 = (0, 0)
    end4 = v

    # Create a figure and axis
    fig, ax = plt.subplots()

    # Plot lines
    #give start,end,label and color for the vector A*u
    ax.plot([start1[0], end1[0]], [start1[1], end1[1]], color='blue', label='A*u')

    #give start,end,label and color for the vector A*v
    ax.plot([start2[0], end2[0]], [start2[1], end2[1]], color='red', label='A*v')

    #give start,end,label and color for the vector u and linestyle to diffenciate
    ax.plot([start3[0], end3[0]], [start3[1], end3[1]], color='blue', linestyle='--', label='u')

    #give start,end,label and color for the vector v and linestyle to diffenciate
    ax.plot([start4[0], end4[0]], [start4[1], end4[1]], color='red', linestyle='--', label='v')

    # Graph labeling of x and y axis
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Visualization of Eigenvector')
    plt.axis('equal')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#transformed vector
transformed_vector_u = mul_fun(A,u)
transformed_vector_v = mul_fun(A,v)
print("u:",transformed_vector_u)
print("v:",transformed_vector_v)
```

the vector u is scaled by value -4 which is eigenvalue, [-24,20]= -4[6,-5](Au = λu)  
while for vector v it is not it is scaled by single constant value.

```python
#define the matrix and vector
A =[[1,6],[5,2]]
u=[6,-5]
v=[3,-2]
plot_graph(A,u,v)
```

As the direction of vector u doesn't changed it is a eigenvector of matrix A while v is not a eigen vector as its direction chaged after transformation
