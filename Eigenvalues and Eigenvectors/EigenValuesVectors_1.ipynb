```python
import matplotlib.pyplot as plt
```

```python
#function to multiplay the matrix and the vector
def mul_fun(x,y):
  if len(x[0])!=len(y):
    raise ValueError("the no of column in matrix and vector lenght should be equal")
  result= []
  for i in range(len(x)):
        sum = 0
        for k in range(len(y)):
            sum += x[i][k] * y[k] #multiplay the elements of the matrix and the vector
        result.append(sum)
  return result
```

```python
#function to plot the graph
def plot_graph(A, u, v):
    start1 = (0, 0)
    end1 = mul_fun(A, u)

    start2 = (0, 0)
    end2 = mul_fun(A, v)

    start3 = (0, 0)
    end3 = u

    start4 = (0, 0)
    end4 = v

    # Create a figure and axis
    fig, ax = plt.subplots()

    # Plot lines
    ax.plot([start1[0], end1[0]], [start1[1], end1[1]], color='blue', label='A*u')
    ax.plot([start2[0], end2[0]], [start2[1], end2[1]], color='red', label='A*v')
    ax.plot([start3[0], end3[0]], [start3[1], end3[1]], color='blue', linestyle='--', label='u')
    ax.plot([start4[0], end4[0]], [start4[1], end4[1]], color='red', linestyle='--', label='v')

    # Graph labeling of x and y axis
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Visualization of Eigenvector')
    plt.axis('equal')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
A =[[1,6],[5,2]]
u=[6,-5]
v=[3,-2]
plot_graph(A,u,v)
```
